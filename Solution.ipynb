```python
import numpy as np
import pandas as pd

# import the original data in dataframe using .read_csv
df = pd.read_csv("data.csv")

# general examination of the entire df
print(df)
df.info()

# examine a certain column
df['geneName']

# examine a certain row
df.iloc[1]
```

```python
# manipulations of dataframe (mix and match basically, select stuffs you need)

# =========================================================================================================================#
# task1: extracting chr info from dataframe and creating a new column to store this info
# explaining this line 
# df['col_name'] showcase, str.replace(r'ori_text', 'new_txt')
# and introduing my best friend google! (learn to describe a problem precisely and concisely ~= solving this problem)
df['chr'] = df['geneName'].str.replace(r'.*_', '')
print(df)
```

```python
# =========================================================================================================================#
# task2: finding the chr with the most amount of genes
# count number of different values in column chr, and reset another column with a default name index for those values

# 2.1: count genes from different chromosomes and construct another dataframe
df_chr_sum = df['chr'].value_counts().reset_index()
df_chr_sum.columns = ['chr', 'count']
print(df_chr_sum)

# 2.2: find the chr with max # of genes
print (df_chr_sum.loc[df_chr_sum['count'] == max(df_chr_sum['count']), 'chr'])

# 2.3: extract the only genes from chr1
df = df[df['chr'] == 'chr1']
print(df)
df.info()
```

```python
# task 4: t-test on significance between OSvPS at the same tp with 2 replicate
# 4.1: sebset data of OSvPS at each tp
df_1hr = df[['geneName', 'setA_OS_1_hr', 'setB_OS_1_hr', 'setA_PS_1_hr', 'setB_PS_1_hr']]
df_1hr.info()
```

```python
# t-test
import numpy as np
from scipy import stats

tp_set = [1, 2, 3, 4, 6, 9, 12, 16, 20, 24]
df_result = pd.DataFrame(columns = ['tp', 't-val', 'p-val']) 

# introducing python for loop, simple type casting, and more df
for i in tp_set:
    tp_A = 'setA_OS_' + str(i) + '_hr'
    tp_B = 'setB_OS_' + str(i) + '_hr'
    t, p = stats.ttest_ind(df[tp_A], df[tp_B])
    df_result = df_result.append({'tp':str(i)+"hr", 't-val':t, 'p-val':p}, ignore_index=True)  

print(df_result)
```
